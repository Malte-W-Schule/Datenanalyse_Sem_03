```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snus

df = pd.read_csv('iris.csv',sep=',')
print(df.head())
```

```python
plt.scatter(df['sepal.length'], df['sepal.width'],marker="+")
plt.xlabel('sepal.length')
plt.ylabel('sepal.width')
plt.show()
```

```python
plt.scatter(df['petal.length'], df['petal.width'],marker="o",color="green")
plt.xlabel('petal.length')
plt.ylabel('petal.width')
plt.show()
```

```python
#b)
cor1 = df["petal.length"].corr(df["petal.width"])
cor2 = df["sepal.length"].corr(df["sepal.width"])

def printCorr(corrwert):
    print(f"The correlation between sepal length and petal.width is {corrwert}")

printCorr(cor1)
printCorr(cor2)
```
