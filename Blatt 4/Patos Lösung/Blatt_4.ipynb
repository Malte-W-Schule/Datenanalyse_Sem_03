```python
from numpy.ma.core import shape
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd

iris = load_iris()
column_names = iris.feature_names
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

setosa = iris.target == 0
df_setosa = df_iris[setosa]

versicolor = iris.target == 1
df_versicolor = df_iris[versicolor]

virginica = iris.target == 2
df_virginica = df_iris[virginica]

plt.hist(df_setosa[column_names[0]], color='blue')
plt.hist(df_versicolor[column_names[0]], color='red')
plt.hist(df_virginica[column_names[0]], color='green')
plt.xlabel(column_names[0])
plt.legend(iris.target_names)
plt.show()

plt.hist(df_setosa[column_names[1]], color='blue')
plt.hist(df_versicolor[column_names[1]], color='red')
plt.hist(df_virginica[column_names[1]], color='green')
plt.xlabel(column_names[1])
plt.legend(iris.target_names)
plt.show()

plt.hist(df_setosa[column_names[2]], color = 'blue')
plt.hist(df_versicolor[column_names[2]], color = 'red')
plt.hist(df_virginica[column_names[2]], color = 'green')
plt.xlabel(column_names[2])
plt.legend(iris.target_names)
plt.show()

plt.hist(df_setosa[column_names[3]], color='blue')
plt.hist(df_versicolor[column_names[3]], color='red')
plt.hist(df_virginica[column_names[3]], color='green')
plt.xlabel(column_names[3])
plt.legend(iris.target_names)
plt.show()
```

Aufgabe 2

```python
#a)

plt.scatter(df_iris[column_names[0]], df_iris[column_names[1]], marker='+')
plt.xlabel(column_names[0])
plt.ylabel(column_names[1])
plt.show()

plt.scatter(df_iris[column_names[2]], df_iris[column_names[3]], marker='o', color='darkgreen')
plt.xlabel(column_names[2])
plt.ylabel(column_names[3])
plt.show()

df_sepal = df_iris[[column_names[0], column_names[1]]]
df_petal = df_iris[[column_names[2], column_names[3]]]
sepal_corr = df_sepal.corr().at[column_names[0], column_names[1]]
petal_corr = df_petal.corr().at[column_names[2], column_names[3]]

print("The correlation coefficient between sepal length (cm) and sepal width (cm) attributes is", sepal_corr, ".")
print("The correlation coefficient between petal length (cm) and petal width (cm) attributes is", petal_corr, ".")
```

```python
#a)
pvtest = pd.read_csv("pvtest.csv")
print(pvtest.head(10))
```

```python
#b)
names = ['Dci', 'Dcp', 'Dcu', 'Temp1']
print(pvtest[names].head(10))
```

```python
#c)
df_select = pvtest[names]
cond1 = pvtest.Dci > 0
cond2 = pvtest.Dcp > 0
cond3 = pvtest.Dcu > 0
cond4 = pvtest.Temp1 > 0

df_select = df_select[cond1 & cond2 & cond3 & cond4]
print(df_select.head(10))
```

```python
#d)
import seaborn as sns
select_corr = df_select.corr()
sns.heatmap(select_corr, annot=True)
```
