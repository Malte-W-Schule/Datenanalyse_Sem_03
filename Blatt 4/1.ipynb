```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snus

df = pd.read_csv('iris.csv',sep=',')
print(df.head())
```

```python
setosa_con = df[df["variety"] == "Setosa"]
versicolor_con = df[df["variety"] == "Versicolor"]
Virginica_con = df[df["variety"]== "Virginica"]
```

```python

sepal_length_setosa = setosa_con["sepal.length"]
sepal_length_versicolor = versicolor_con["sepal.length"]
sepal_length_virginica = Virginica_con["sepal.length"]

print(sepal_length_setosa.head())
plt.hist(sepal_length_setosa, color="blue", label="Setosa")
plt.hist(sepal_length_versicolor, color="red", label="Versicolor")
plt.hist(sepal_length_virginica, color="green", label="Virginica")

plt.legend()
plt.show()
```
