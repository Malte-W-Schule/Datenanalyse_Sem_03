```python
#c)

werte = df[["Dci","Dcp","Dcu","Temp1"]]

cond1 = werte.Dci > 0
cond2 = werte.Dcp > 0
cond3 = werte.Dcu > 0
cond4 = werte.Temp1 > 0

werte_constrained = werte[cond1 & cond2 & cond3 & cond4]
print(werte_constrained.head(10))
```
