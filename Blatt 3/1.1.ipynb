```python
import pandas as pd
import re

data = []

with open('countries_population.csv', 'r', encoding='utf-8') as f:
    for line in f:
        # Mit Regex: alles zwischen den Hochkommas als Land, Rest als Population
        match = re.match(r"'(.+?)'\s+([\d,]+)", line.strip())
        if match:
            country = match.group(1)            # LÃ¤ndername ohne Hochkommas
            population = int(match.group(2).replace(',', ''))  # Zahl ohne Tausender-Kommas
            data.append([country, population])

# DataFrame erzeugen
df = pd.DataFrame(data, columns=['Country', 'Population'])
df.set_index('Country', inplace=True)

print(df.head())
```
