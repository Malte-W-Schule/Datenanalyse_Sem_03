```python
#a)

fname2 = "bundeslaender.txt"
column_names = ['land', 'area', 'female', 'male', 'population', 'density']
bundeslaender = pd.read_csv(fname2, sep=" ", quotechar="'", index_col=0, header=0)
bundeslaender['population'] = bundeslaender['female'] + bundeslaender['male']
bundeslaender['density'] = (bundeslaender['population'] / bundeslaender['area']) * 1000

print(bundeslaender)
```

```python
#b)

cond1a = bundeslaender['female']
cond1b = bundeslaender['male']
print(bundeslaender[cond1a > cond1b])
```

```python
cond2 = bundeslaender['density'] > 1000
print(bundeslaender[cond2])
```

```python
#a)

index = pd.date_range('12/24/2025', '1/6/2026')
weekdays = index[index.weekday < 5]

print(weekdays)
print("Es gibt", len(weekdays), "normale Wochentage/ reguläre Arbeitstage in den Weihnachtsferien 2025/26.")
```

```python
#b)

index2 = pd.date_range('12/24/2025', '1/6/2027')
cond3a = index2.weekday == 6
cond3b = index2.day == 1
sundays = index2[cond3a & cond3b]

print(sundays)
print("Vom 24.12.2025 bis zum 6.01.2027 gibt es", len(sundays), "Sonntage, die auf dem 1. des Monats fallen.")
```
